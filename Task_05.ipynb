```python
import pandas as pd
df = pd.read_csv('/content/loan_dataset - Copy.csv')
df.head(3)
```

```python
df.dtypes
```

```python
# find duplicates and print
duplicates = df[df.duplicated()]
duplicates
```

```python
# drop duplicates rows
df = df.drop_duplicates()
```

```python
# droping NaN values :
df.dropna(subset=df.columns, inplace=True)
```

```python
# find columns with dtypes as object
df1 = df.select_dtypes(include=['object'])
```

```python
for i in df1.columns:
  print(i,':',df[i].unique())
```

```python
# Check for unique values in each column
unique_counts = df1.nunique()

# Find columns with all unique values
columns_to_drop = unique_counts[unique_counts == df1.shape[0]].index

# Drop columns with all unique values
df1 = df1.drop(columns=columns_to_drop)
df1
```

```python

# Create a list to store the column names, unique values, and counts
columns_with_unique_values = []

for column_name in df1.columns:
    unique_values = df1[column_name].unique()
    counts = [df1[column_name].value_counts()[value] for value in unique_values]

    columns_with_unique_values.append({
        'Column Name': column_name,
        'Unique Values': unique_values,
        'Counts of U.Values': counts
    })

# Create a new DataFrame
columns_and_values_df = pd.DataFrame(columns_with_unique_values)

print( 'CHOOSE THE COLUMN TO STANDARDIZE :\n',columns_and_values_df)
```

```python
import pandas as pd



numeric_columns = []

# Function to check if a value can be converted to numeric
def is_numeric(x):
    try:
        # Attempt to convert the value to numeric, considering '+' as valid
        float(x.replace('+', ''))
        return True
    except (ValueError, AttributeError):
        return False

# Loop through the columns and check if all values can be converted to numeric
for col_name in df1.columns:
    if df1[col_name].apply(is_numeric).all():
        numeric_columns.append(col_name)


print("Columns with 'numeric' data but dtype as 'object':")
print(numeric_columns)
```
