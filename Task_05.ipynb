```python
import pandas as pd
df = pd.read_csv('/content/loan_dataset - Copy.csv')
df.head(3)
```

```python
df.dtypes
```

```python
# find duplicates and print
duplicates = df[df.duplicated()]
duplicates
```

```python
# drop duplicates rows
df = df.drop_duplicates()
```

```python
# droping NaN values :
df.dropna(subset=df.columns, inplace=True)
```

```python
# find columns with dtypes as object
df1 = df.select_dtypes(include=['object'])
```

```python
for i in df1.columns:
  print(i,':',df[i].unique())
```

```python
df1.dtypes

```

```python
df1
```

```python
# Check for unique values in each column
unique_counts = df1.nunique()

# Find columns with all unique values
columns_to_drop = unique_counts[unique_counts == df1.shape[0]].index

# Drop columns with all unique values
df1 = df1.drop(columns=columns_to_drop)
df1
```

```python
df1.dtypes
```

```python
import pandas as pd

# Specify the names of the columns you want to standardize
columns_to_standardize = df1.columns
# Add the names of the columns you want to standardize to this list

# Create a DataFrame to store the standardized data
standardized_df = df.copy()  # Start with a copy of the original DataFrame

# Iterate through the specified columns
for column_name in columns_to_standardize:
    if column_name in df.columns:
        # Process only the specified columns
        standardized_df[column_name] = standardized_df[column_name].str.capitalize()

for column_name in columns_to_standardize:
    if column_name in standardized_df.columns:
        # Define a function to convert single letters to the mode value
          def convert_to_mode(group):
              mode_value = group.mode().iloc[0]  # Calculate the mode of the group
              return mode_value

          # Apply the function to each group of values with the same starting letter
          standardized_df[column_name] = standardized_df.groupby(standardized_df[column_name].str[0])[column_name].transform(convert_to_mode)


# Now, the 'standardized_df' DataFrame contains the capitalized values for the specified columns.
standardized_df.head(7)
```
